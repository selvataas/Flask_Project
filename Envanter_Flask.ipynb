```python
import sqlite3

def create_tables():
    conn = sqlite3.connect('inventory.db')
    cursor = conn.cursor()
    
    cursor.execute('''
    CREATE TABLE IF NOT EXISTS YN (
        YNid INTEGER PRIMARY KEY,
        YNTur VARCHAR(10),
        YNAdres VARCHAR(20)
    )
    ''')
    
    cursor.execute('''
    CREATE TABLE IF NOT EXISTS HN (
        HNid INTEGER PRIMARY KEY,
        HNTur VARCHAR(10),
        HNAdres VARCHAR(20)
    )
    ''')
    
    cursor.execute('''
    CREATE TABLE IF NOT EXISTS Mesafe (
        YN_YNid INTEGER,
        HN_HNid INTEGER,
        M INTEGER,
        FOREIGN KEY (YN_YNid) REFERENCES YN(YNid),
        FOREIGN KEY (HN_HNid) REFERENCES HN(HNid)
    )
    ''')
    
    cursor.execute('''
    CREATE TABLE IF NOT EXISTS Ekipman (
        Eid INTEGER PRIMARY KEY,
        ETur VARCHAR(20),
        EPlaka VARCHAR(10),
        EKapasite INTEGER,
        EBosHiz INTEGER,
        EDoluHiz INTEGER,
        EManervaSuresi INTEGER,
        EUygunluk INTEGER
    )
    ''')

#         EBosaltmaSuresi INTEGER,
#         EYuklemeSuresi INTEGER,
    
    cursor.execute('''
    CREATE TABLE IF NOT EXISTS Kullanici (
        Kid INTEGER PRIMARY KEY,
        KAdi VARCHAR(45),
        KSoyadi VARCHAR(45),
        S_Sid INTEGER,
        FOREIGN KEY (S_Sid) REFERENCES Sorunlar(Sid)
    )
    ''')
    
    cursor.execute('''
    CREATE TABLE IF NOT EXISTS Sorunlar (
        Sid INTEGER PRIMARY KEY,
        SorunAciklama VARCHAR(50)
    )
    ''')
    
    cursor.execute('''
    CREATE TABLE IF NOT EXISTS EK (
        E_Eid INTEGER,
        K_Kid INTEGER,
        FOREIGN KEY (E_Eid) REFERENCES Ekipman(Eid),
        FOREIGN KEY (K_Kid) REFERENCES Kullanici(Kid)
    )
    ''')
    
    cursor.execute('''
    CREATE TABLE IF NOT EXISTS HNE (
        E_Eid INTEGER,
        HN_HNid INTEGER,
        FOREIGN KEY (E_Eid) REFERENCES Ekipman(Eid),
        FOREIGN KEY (HN_HNid) REFERENCES HN(HNid)
    )
    ''')
    
    cursor.execute('''
    CREATE TABLE IF NOT EXISTS YNE (
        YN_YNid INTEGER,
        E_Eid INTEGER,
        FOREIGN KEY (YN_YNid) REFERENCES YN(YNid),
        FOREIGN KEY (E_Eid) REFERENCES Ekipman(Eid)
    )
    ''')

    conn.commit()
    conn.close()

create_tables()
```

```python
from random import randint

def add_distance_data():
    with sqlite3.connect('inventory.db') as conn:
        cursor = conn.cursor()
        cursor.execute('SELECT YNid FROM YN')
        yukleme_noktalari = cursor.fetchall()
        cursor.execute('SELECT HNid FROM HN')
        hedef_noktalari = cursor.fetchall()
        # YN ve HN tablolarındaki her bir kombinasyon için rastgele bir mesafe oluşturup Mesafe tablosuna ekledim.
        for yn in yukleme_noktalari:
            for hn in hedef_noktalari:
                mesafe = randint(1, 10)  
                cursor.execute('INSERT INTO Mesafe (YN_YNid, HN_HNid, M) VALUES (?, ?, ?)', (yn[0], hn[0], mesafe))

add_distance_data()
print("Mesafe verileri başarıyla eklendi.")
```

```python
import sqlite3

# Veritabanı bağlantısını oluştur
with sqlite3.connect('inventory.db') as conn:
    cursor = conn.cursor()

    # Tüm tabloları al
    cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
    tables = cursor.fetchall()

    # Her tablo için içeriği görüntüle
    for table in tables:
        table_name = table[0]
        print(f"Table: {table_name}")
        cursor.execute(f"SELECT * FROM {table_name}")
        rows = cursor.fetchall()
        for row in rows:
            print(row)
        print("\n")
```

```python
from flask import Flask, render_template, request, redirect, url_for, session
import sqlite3

app = Flask(__name__)
app.secret_key = '987654321'

def get_db_connection():
    conn = sqlite3.connect('inventory.db')
    conn.row_factory = sqlite3.Row
    return conn

@app.route('/')
def index():
    return render_template('index.html')

@app.route('/login', methods=['GET', 'POST'])
def login():
    if request.method == 'POST':
        username = request.form['username']
        password = request.form['password']
        conn = get_db_connection()
        user = conn.execute('SELECT * FROM Kullanici WHERE KAdi = ? AND KSoyadi = ?', (username, password)).fetchone()
        conn.close()
        if user:
            session['user_id'] = user['Kid']
            return redirect(url_for('dashboard'))
        else:
            return 'Invalid username or password'
    return render_template('login.html')

@app.route('/register', methods=['GET', 'POST'])
def register():
    if request.method == 'POST':
        username = request.form['username']
        password = request.form['password']
        conn = get_db_connection()
        conn.execute('INSERT INTO Kullanici (KAdi, KSoyadi) VALUES (?, ?)', (username, password))
        conn.commit()
        conn.close()
        return redirect(url_for('login'))
    return render_template('register.html')

@app.route('/dashboard')
def dashboard():
    if 'user_id' not in session:
        return redirect(url_for('login'))
    
    conn = get_db_connection()
    ekipmanlar = conn.execute('SELECT * FROM Ekipman').fetchall()
    yukleme_noktalari = conn.execute('SELECT * FROM YN').fetchall()
    hedef_noktalari = conn.execute('SELECT * FROM HN').fetchall()
    conn.close()
    return render_template('dashboard.html', ekipmanlar=ekipmanlar, yukleme_noktalari=yukleme_noktalari, hedef_noktalari=hedef_noktalari)

@app.route('/transfer', methods=['POST'])
def transfer():
    if 'user_id' not in session:
        return redirect(url_for('login'))
    
    ekipman_id = request.form['ekipman_id']
    yukleme_noktasi = request.form['yukleme_noktasi']
    hedef_noktasi = request.form['hedef_noktasi']
    
    conn = get_db_connection()
    conn.execute('INSERT INTO YNE (YN_YNid, E_Eid) VALUES (?, ?)', (yukleme_noktasi, ekipman_id))
    conn.execute('INSERT INTO HNE (HN_HNid, E_Eid) VALUES (?, ?)', (hedef_noktasi, ekipman_id))
    conn.commit()
    conn.close()
    
    return redirect(url_for('dashboard'))

@app.route('/logout')
def logout():
    session.pop('user_id', None)
    return redirect(url_for('index'))


if __name__ == '__main__':
    app.run(debug=False, port=8001)
```
